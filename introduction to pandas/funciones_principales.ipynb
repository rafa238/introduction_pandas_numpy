# Funciones principales de pandas

```python
import pandas as pd
```

```python
df_books = pd.read_csv("./bestsellers-with-categories.csv", sep=',', header=0)
df_books.head(2)
```

### Info

Con esta función podemos obtener un pequeño resumen de las columnas, los datos no nulos por columna y el tipo de dato de la columna.

```python
df_books.info()
```

### Describe

De las columnas númericas podemos obtener datos estadisticos basiccos de cada columna numerica.

```python
df_books.describe()
```

### Head y Tail

Estas funciones devuelven los primeros o ultimos registros del dataset.

```python
df_books.tail(2)
```

### Memory usage

Devuelve la cantidad de memoria que utiliza cada columna.

```python
df_books.memory_usage(deep = True)
```

### Value Counts

Con esta función podemos ver datos repetidos y cuantas veces.

```python
df_books['Author'].value_counts()
```

### Drop duplicates

Esta funcion se encarga de borrar filas repetidas

```python
df_books.append(df_books.iloc[-10:, ])
```

```python
df_books.drop_duplicates()
```

### Sort values

Para este caso vamos a ordenar valores por columna

```python
df_books.sort_values("Year")
```

Pero podemos ordenarla a la inversa

```python
df_books.sort_values("Year", ascending=False)
```

## Group By

Esta funcion funciona exactamente como en SQL, agrupo valores y nos dice informacion sobre este conjunto agrupado.

Veamos la meadia.

```python
df_books.groupby("Author").mean()
```

En este caso vimos que nos dice la media de todas las columnas numericas de los autores agrupados.

Ahora contemos todos los libros agrupados por genero y veamos cuantos datos validos tiene por columna.

```python
df_books.groupby("Genre").count()
```

Groupby convierte nuestro dataframe a un modelo en donde el parametro que le pasamos al group se convierte
en el indice, y por lo tanto, podemos acceder a un elemento con su label (loc)

```python
df_books.groupby("Author").sum().loc["Walter Isaacson"]
```

Para tener nuestro dataset agrupado y ocupar el indice original, utilizamos reset index

```python
df_books.groupby("Author").sum().reset_index()
```

Ahora agruparemos pero tomaremos ciertas funciones en cada columna
Osea, aplicar mas de una funcion al agrupamiento por cada columna.

```python
df_books.groupby("Author").agg(["min", "max"])
```

Ahora sera lo mismo, pero seleccionaremos columnas

```python
df_books.groupby("Author").agg(
    {
        "User Rating": ["min", "max"],
        "Reviews": ["count"]
    }
)
```

Tambien multiples agrupamientos

```python
df_books.groupby(["Author", "Year"]).count()
```
