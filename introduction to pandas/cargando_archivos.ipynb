# Leyendo archivos con pandas
En esta seccion vamos cargar archivos json y csv para cargar su información en un dataFram de pandas, mismos que son los mas comunes para tratar información en ciencia de datos.

```python
import pandas as pd
```

En este caso vamos a cargar el archivo que contiene los best seller de una coleccion de libros.

El parametro sep sera el separador por el cual el archivo esta separando el contenido de las filas, en este caso csv es una coma (pandas utiliza , por defecto).

El parametro header es la fila que utilzaremos como nombres de las columnas.

Names, es una lista donde renombraremos a las columnas que queremos nombrar.

```python
df_books = pd.read_csv('./bestsellers-with-categories.csv', sep=',', header=0, names=None)
```

# Loc
Loc nos permite navegar por el dataframe teniendo como primer parametro indices de filas (slicing o single), y en el segundo parametro nos permite tratar con las columnas por nombre (lista de strings o un solo string que represente una sola columna).

Primero veamos como se haria sin Loc.

Podemos hacer slicing y obtener una cantidad limitada de registros.

```python
df_books[0:2]
```

Podemos filtrar por columna con su nombre

```python
df_books['Name']
```

Por mas de 1 columna

```python
df_books[['Name', 'Author']]
```

¿Que pasa si queremos filtrar de tal manera que tomemos una cantidad limitadaa de registros (como previamente lo hicimos con slicing), pero tambien queremos tomar una cantidad limitada de columnas?

La función loc viene a resolver nuestros problemas :D
Es por ello que ella nos va a permitir combinar ambas operaciones (slicing y selección de columnas).

```python
df_books.loc[0:4, ['Name', 'Author']]
```

Podemos manejar directamente los labels o elementos de una columna y realizar operaciones y condiciones sobre ellas

```python
df_books.loc[:, ['Author']] == 'JJ Smith'
```

## ILOC
iloc nos permite trabajar de manera similar a loc pero el segundo y primer parametro nos permitira navegar en las columnas y filas por indices, a diferencia de loc que solo lo permite por nombres.

En este caso, tendremos todos los registros pero de la 4ta columna en adelante.

```python
df_books.iloc[:, 4:]
```
